# Selecting canonical_smiles and molecule_chembl_id

```python
selection = ['canonical_smiles','molecule_chembl_id']
df_selection = df[selection]
df_selection.to_csv('molecule.smi', sep='\t', index=False, header=False)
df_selection
```

# Calculate descriptor

```python
padeldescriptor(
    mol_dir="molecule.smi",       # input file
    d_file="descriptors.csv",      # output file
    d_2d=False,                     # compute 2D descriptors
    fingerprints=True,             # compute fingerprints
    threads=1,                     # use 4 CPU threads
    removesalt=True,
    standardizenitro=True,
    headless=True     # no splash screen
)
```
