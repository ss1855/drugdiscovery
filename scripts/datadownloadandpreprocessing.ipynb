# Import necessary libraries

```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
import os
```

# Search the chembl database for target organism/compound
Returns the search result as dictionary. Convert it into dictionary

```python
target = new_client.target
target_query = target.search('acetylcholinesterase')
targets = pd.DataFrame.from_dict(target_query)
targets
```

# Select the target you are interested in

```python
selected_target = targets.target_chembl_id[1]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id = selected_target).filter(standard_type = "IC50")
```

```python
print(f"Number of records: {len(res)}")
print("Sample record keys:", res [0].keys())
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df
```

```python

os.makedirs("data", exist_ok=True)  # Create folder if it doesn't already exist
df.to_csv("data/bioactivity_data_raw.csv", index=False)
```

# Data preprocessing

## Handling Missing data and removing duplicate canonical_smiles
If any compounds has missing value for the standard_value  and canonical smiles column then drop it


```python
df2 = df[df.standard_value.notna()]
df2 = df2[df2.canonical_smiles.notna()]
```

```python
df2.standard_value.min()
```

```python
len(df2.canonical_smiles.unique())
```

```python
duplicates = df2[df2.duplicated(subset='canonical_smiles', keep=False)][
    ['canonical_smiles', 'molecule_chembl_id', 'standard_value']
]
```

```python
duplicates
```

# Check and remove any canonical duplicates and remove duplicate entry keeping only the first appearance in dataframe

```python
df2_nr = df2.drop_duplicates(['canonical_smiles'])
df2_nr
```

### **Combine the 3 columns (molecule_chembl_id,canonical_smiles,standard_value) and bioactivity_class into a DataFrame**

```python
selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df3 = df2_nr[selection]
df3 = df3.reset_index(drop=True)
df3.to_csv("data/bioactivity_data_preprocessed.csv", index=False)
```

## Data Labelling

```python
bioactivity_class = []
for i in df3.standard_value:
    if float(i) >= 10000:
        bioactivity_class.append("Inactive")
    elif float(i) <= 1000:
        bioactivity_class.append("Active")
    else:
        bioactivity_class.append("Intermediate")
```

```python
len(bioactivity_class)
```

```python
bioactivity_class = pd.Series(bioactivity_class, name='bioactivity_class')
```

```python
bioactivity_class
```

```python
df4 = pd.concat([df3,bioactivity_class],axis =1 )
```

```python
df4.to_csv("data/bioactivity_data_curated.csv", index=False)
```
